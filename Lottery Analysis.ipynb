```python
import random


def lotto_game():
    #Initialise an empty list that will be used to store the 5 lucky numbers!
    LottoNumbers = []
    for i in range (0,6):
        number = random.randint(1,59)
        #Check if this number has already been picked and ...
        while number in LottoNumbers:
            # ... if it has, pick a new number instead 
            number = random.randint(1,59)
        #Now that we have a unique number, let's append it to our list.
        LottoNumbers.append(number)
    #Sort the list in ascending order
    #LottoNumbers.sort()
    return LottoNumbers

def lotto_draw_simulate(n):
    #Initialise an empty list for each draw
    Draw = []
    #For each draw...
    for i in range(n):
        #...Append the result of each loto game
        Draw.append(lotto_game())
    return Draw

#Specify number of draws and save output to a variable
lotto_batch_results = lotto_draw_simulate(700)
#print(lotto_batch_results)
```

```python
import pandas as pd


#Convert simuated game results into a dataframe
final_lotto_df = pd.DataFrame(lotto_batch_results, columns=["1st Ball", "2nd Ball", "3rd Ball", "4th Ball", "5th Ball", "6th Ball"])
print(final_lotto_df)
```

```python
#Record frequency of repeat draw results per ball, as new dataframes
first_count = final_lotto_df['1st Ball'].value_counts()
first_df = pd.DataFrame(first_count).reset_index()
first_df.columns = ['Ball', 'Count']
#print(first_df)
second_count = final_lotto_df['2nd Ball'].value_counts()
second_df = pd.DataFrame(second_count).reset_index()
second_df.columns = ['Ball', 'Count']
#print(second_df)
third_count = final_lotto_df['3rd Ball'].value_counts()
third_df = pd.DataFrame(third_count).reset_index()
third_df.columns = ['Ball', 'Count']
#print(third_df)
fourth_count = final_lotto_df['4th Ball'].value_counts()
fourth_df = pd.DataFrame(fourth_count).reset_index()
fourth_df.columns = ['Ball', 'Count']
#print(fourth_df)
fifth_count = final_lotto_df['5th Ball'].value_counts()
fifth_df = pd.DataFrame(fifth_count).reset_index()
fifth_df.columns = ['Ball', 'Count']
#print(fifth_df)
sixth_count = final_lotto_df['6th Ball'].value_counts()
sixth_df = pd.DataFrame(sixth_count).reset_index()
sixth_df.columns = ['Ball', 'Count']
#print(sixth_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 


sns.set_style('whitegrid')
sns.lmplot(x ='Ball', y ='Count', data = first_df).set(title='1st Ball')
plt.show()
sns.lmplot(x ='Ball', y ='Count', data = second_df).set(title='2nd Ball')
plt.show()
sns.lmplot(x ='Ball', y ='Count', data = third_df).set(title='3rd Ball')
plt.show()
sns.lmplot(x ='Ball', y ='Count', data = fourth_df).set(title='4th Ball')
plt.show()
sns.lmplot(x ='Ball', y ='Count', data = fifth_df).set(title='5th Ball')
plt.show()
sns.lmplot(x ='Ball', y ='Count', data = sixth_df).set(title='6th Ball')
plt.show()
```
